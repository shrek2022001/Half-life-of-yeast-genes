```python
import numpy as np
from sklearn.linear_model import LinearRegression

half_lives_1 = {}

# Iterate over each gene
for gene in genes_1:
    # Extract expression levels for the gene
    expression_levels = replicate1_transposed[gene].values
    
    # Ensure all values are numeric
    expression_levels = pd.to_numeric(expression_levels, errors='coerce')
    
    # Remove NaN values
    expression_levels = expression_levels[~np.isnan(expression_levels)]
    
    # Take the natural logarithm of expression levels
    log_expression = np.log(expression_levels)
    
    # Check for -inf or inf values after taking the logarithm
    if np.any(np.isinf(log_expression)):
        print(f"Warning: Log transformation resulted in infinite values for gene {gene}. Skipping.")
        continue
    
    # Remove -inf or inf values
    log_expression = log_expression[np.isfinite(log_expression)]
    
    # Fit a linear regression model
    model = LinearRegression()
    
    # Check if log_expression still contains values after removing -inf or inf
    if len(log_expression) > 0:
        # Fit the model if log_expression is not empty
        model.fit(time_points[:len(log_expression)], log_expression)
        
        # Extract slope of the linear fit, which corresponds to decay rate (λ)
        decay_rate = model.coef_[0]
        
        # Calculate half-life (t1/2)
        half_life = -np.log(2) / decay_rate
        
        # Store half-life for the gene
        half_lives_1[gene] = half_life
    else:
        print(f"No valid data for gene {gene}")

# Print half-lives
for gene, half_life in half_lives_1.items():
    print(f"{gene}: {half_life:.2f} minutes")
```

```python
for gene in full_data_transposed.keys():
    # Get half-lives from each replicate, replace NaN with 0
    half_life_1 = half_lives_1.get(gene, 0)
    half_life_2 = half_lives_2.get(gene, 0)
    half_life_3 = half_lives_3.get(gene, 0)
    
    # Calculate average half-life
    avg_half_life = np.mean([half_life_1, half_life_2, half_life_3])
    
    # Store average half-life in dictionary
    avg_half_lives[gene] = avg_half_life

# Print or use avg_half_lives dictionary as needed
print(avg_half_lives)
```

```python
# Sort the average half-lives dictionary by values in ascending order
sorted_avg_half_lives = dict(sorted(avg_half_lives.items(), key=lambda item: item[1]))

# Calculate the number of genes that correspond to the top 10%
top_10_percent = int(len(sorted_avg_half_lives) * 0.1)

# Open a text file for writing
with open("top_genes.txt", "w") as file:
    count = 0
    for gene, half_life in sorted_avg_half_lives.items():
        if half_life != float('-inf'):
            # Write the gene name to the text file
            file.write(f"{gene}\n")
            count += 1
            if count >= top_10_percent:
                break
```

```python
# Sort the average half-lives dictionary by values in descending order
sorted_avg_half_lives = dict(sorted(avg_half_lives.items(), key=lambda item: item[1], reverse=True))

# Calculate the number of genes that correspond to the bottom 10%
bottom_10_percent = int(len(sorted_avg_half_lives) * 0.1)

# Open a text file for writing
with open("bottom_genes.txt", "w") as file:
    count = 0
    for gene, half_life in sorted_avg_half_lives.items():
        if half_life != float('-inf'):
            # Write the gene name to the text file
            file.write(f"{gene}\n")
            count += 1
            if count >= bottom_10_percent:
                break
```
